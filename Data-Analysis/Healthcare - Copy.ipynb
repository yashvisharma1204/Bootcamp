## Handling null values

```python
def plot_missing_values(df, title):
    missing_values = df.isnull().sum()
    plt.figure(figsize=(16, 10))
    sns.barplot(x=missing_values.index, y=missing_values.values)
    plt.xticks(rotation=90)
    plt.title(title)
    plt.xlabel('Columns')
    plt.ylabel('Number of Missing Values')
    plt.show()
```

```python
plot_missing_values(census_df,"Missing Values in Census DataFrame Before Imputation")
census_df.fillna(census_df.select_dtypes(include=np.number).mean(), inplace=True)
plot_missing_values(census_df, "Missing Values in Census DataFrame After Imputation")
```

```python
plot_missing_values(hospitals_df, "Missing Values in Hospitals DataFrame Before Imputation")
hospitals_df.fillna(hospitals_df.select_dtypes(include=np.number).mean(), inplace=True)
plot_missing_values(hospitals_df, "Missing Values in Hospitals DataFrame After Imputation")
```

## Handling outliers

```python
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        df = df[~((df[col] < (Q1 - 1.5 * IQR)) |(df[col] > (Q3 + 1.5 * IQR)))]
    return df
```

```python
plt.figure(figsize=(16, 10))
sns.boxplot(data=census_df[['Household_size_1_person_Households', 'Household_size_2_persons_Households',
                            'Household_size_3_to_5_persons_Households', 'Household_size_6_8_persons_Households']])
plt.title('Boxplot of Household Sizes')
plt.xlabel('Household Sizes')
plt.ylabel('Count')
plt.show()
census_df_clean = remove_outliers(census_df, ['Household_size_1_person_Households', 
                                                'Household_size_2_persons_Households',
                                                'Household_size_3_to_5_persons_Households',
                                                'Household_size_6_8_persons_Households'])

plt.figure(figsize=(16, 10))
sns.boxplot(data=census_df_clean[['Household_size_1_person_Households', 
                                    'Household_size_2_persons_Households',
                                    'Household_size_3_to_5_persons_Households',
                                    'Household_size_6_8_persons_Households']])
plt.title('Boxplot of Household Sizes (Outliers Removed)')
plt.xlabel('Household Sizes')
plt.ylabel('Count')
plt.show()
```

## Exploratory Data Analysis (EDA) and Visualizations
### Distribution of Population

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(census_df['Population'],label ='Population',color='indianred')
sns.kdeplot(census_df['Male'],label ='Male',color='royalblue')
sns.kdeplot(census_df['Female'],label ='Female',color='mediumaquamarine')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Population Distribution Analysis:** The histogram illustrates the frequency distribution of population sizes in the census dataset, with population values ranging from 0 to 10 million. The majority of the data points are concentrated at lower population values, indicating a right-skewed distribution. The kernel density estimate (KDE) line further highlights this skewness, showing a peak around the lower population range and tapering off towards the higher population values
